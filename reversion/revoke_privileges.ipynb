#  Project Reversion

## Revoke Privileges
Revoke privileges to groups and users assigned to this project

```sql
%sql
-- ---------------------------
-- Catalog-Level Revokes
-- ---------------------------
REVOKE USE CATALOG ON CATALOG ap FROM `data_engineers`;
REVOKE USE CATALOG ON CATALOG ap FROM `analytics_team`;



-- ---------------------------
-- Schema-Level Revokes (ap.gold)
-- ---------------------------
-- DATA ENGINEERS
REVOKE USE SCHEMA ON SCHEMA ap.gold FROM `data_engineers`;
REVOKE SELECT ON SCHEMA ap.gold FROM `data_engineers`;
REVOKE MODIFY ON SCHEMA ap.gold FROM `data_engineers`;
REVOKE CREATE TABLE ON SCHEMA ap.gold FROM `data_engineers`;
REVOKE CREATE VIEW ON SCHEMA ap.gold FROM `data_engineers`;
REVOKE CREATE FUNCTION ON SCHEMA ap.gold FROM `data_engineers`;
REVOKE CREATE MODEL ON SCHEMA ap.gold FROM `data_engineers`;
REVOKE CREATE MATERIALIZED VIEW ON SCHEMA ap.gold FROM `data_engineers`;
REVOKE CREATE VOLUME ON SCHEMA ap.gold FROM `data_engineers`;
REVOKE WRITE VOLUME ON SCHEMA ap.gold FROM `data_engineers`;
REVOKE APPLY TAG ON SCHEMA ap.gold FROM `data_engineers`;

-- ANALYTICS TEAM
REVOKE USE SCHEMA ON SCHEMA ap.gold FROM `analytics_team`;
REVOKE SELECT ON SCHEMA ap.gold FROM `analytics_team`;
REVOKE READ VOLUME ON SCHEMA ap.gold FROM `analytics_team`;

-- EVER (special user)
REVOKE MODIFY ON SCHEMA ap.gold FROM `ever@smartdatacorp.com`;
REVOKE CREATE TABLE ON SCHEMA ap.gold FROM `ever@smartdatacorp.com`;
REVOKE CREATE VIEW ON SCHEMA ap.gold FROM `ever@smartdatacorp.com`;
REVOKE ALL PRIVILEGES ON SCHEMA ap.gold FROM `ever@smartdatacorp.com`;



-- ---------------------------
-- Table-Level Revokes
-- ---------------------------

REVOKE SELECT ON TABLE ap.silver.ap_invoices FROM `analytics_team`;
```
