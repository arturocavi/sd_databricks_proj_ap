## Clean Storage Data
Remove folders and files from storage

```python
# Tables to delete
tables = [
  "ap_invoices",
  "suppliers",
  "gl_control_totals"
]

# Delete tables and checkpoints stored in ADLS
for table in tables:
  table_path = f"abfss://bronze@sd0212.dfs.core.windows.net/{table}"
  checkpoint_path = f"abfss://bronze@sd0212.dfs.core.windows.net/checkpoint_{table}"
  try:
    dbutils.fs.rm(table_path, True)
  except Exception as e:
    print(f"Table path not found or could not be deleted: {table_path} - {e}")
  try:
    dbutils.fs.rm(checkpoint_path, True)
  except Exception as e:
    print(f"Checkpoint path not found or could not be deleted: {checkpoint_path} - {e}")

```
