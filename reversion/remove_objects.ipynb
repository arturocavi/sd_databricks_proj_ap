## Drop Catalog
Dropping all schemas and tables in it

```python
# Remove previous widgets
dbutils.widgets.removeAll()
```

```python
# Create the parameter
dbutils.widgets.text("catalog_name", "ap_dev")

#Read the parameter
catalog_name = dbutils.widgets.get("catalog_name")
```

```python
spark.sql(f"DROP CATALOG IF EXISTS {catalog_name} CASCADE;")
```

## Clean Storage Data
Remove folders and files from storage

```python
# Tables to delete
tables = [
  "ap_invoices",
  "suppliers",
  "gl_control_totals"
]

# Delete tables and checkpoints stored in ADLS
for table in tables:
  table_path = f"abfss://bronze@sd0212.dfs.core.windows.net/{table}"
  checkpoint_path = f"abfss://bronze@sd0212.dfs.core.windows.net/checkpoint_{table}"
  try:
    dbutils.fs.rm(table_path, True)
  except Exception as e:
    print(f"Table path not found or could not be deleted: {table_path} - {e}")
  try:
    dbutils.fs.rm(checkpoint_path, True)
  except Exception as e:
    print(f"Checkpoint path not found or could not be deleted: {checkpoint_path} - {e}")

```
